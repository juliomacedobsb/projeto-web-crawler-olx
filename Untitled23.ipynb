na primeira e segunda linha, estamos importando as bibliotecas Beautiful Soup, Requests e a biblioteca Pandas, que é uma biblioteca de visualização de dados em forma de tabela

```python
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd
```

Indicamos o headers que passa informações adicionais a requisição requests, nesse caso está informando o usuário que está fazendo a requisição.  Na segunda linha, informamos a url que queremos fazer o web scraping e na terceira linha, pegamos o conteúdo da requisição GET na url de olx.com/imóveis/aluguel, e nas próximas é criado o objeto soup que interpreta o documento Html.

```python
headers = ({'User-Agent' : 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:102.0) Gecko/20100101 Firefox/102.0'})
url = 'https://df.olx.com.br/imoveis/aluguel'
response = rq.get(url,  headers=headers)
content = response.content

site = bs(content, 'html.parser')

anuncios = site.find_all("ul", class_='sc-1fcmfeb-1 kntIvV')
```

Utilizando a função FIND do Beautiful Soup procuramos pelos dados da primeira html solicitada, e utilizando o FIND_ALL percorreremos por todos os anúncios de aluguel de imóveis olx, criamos então listas e adicionamos os dados selecionados para a extração de maneira organizada e mais limpa possível, no nosso projeto pesquisamos, titulo do anuncio, valor do aluguel, link , localização, data da postagem, metragem do imóvel, quantidade de quartos e valor do condomínio.

```python
titulos = []
links = []
valores = []

itens = site.find('ul', id='ad-list')
for item in itens.find_all('a'):
    titulos.append(item['title'])
    links.append(item['href'])
```

```python
itens = site.find('ul', id='ad-list')
for item in itens.find_all('div', class_='sc-1kn4z61-1 dGMPPn'):
    valores.append(item.text)
    if item.text:
        valores.append(item.text)
    else:
        valores.append('')
```

```python
localizacao = []


itens = site.find('ul', id='ad-list')

for item in itens.find_all('span', class_='sc-1c3ysll-1 iDvjkv sc-fzsDOv dTHJIA'):
    localizacao.append(item['title'])
    if item.text:
        localizacao.append(item.text)
    else:
        localizacao.append('')
```

```python
data = []


itens = site.find('ul', id='ad-list')

for item in itens.find_all('span', class_='sc-11h4wdr-0 javKJU sc-fzsDOv dTHJIA'):
    data.append(item['aria-label'])
    if item.text:
        data.append(item.text)
    else:
        data.append('')
```

```python
metros = []
quartos = []
cond = []

itens = site.find('ul', id='ad-list')
for item in itens.find_all('span', class_='sc-1ftm7qz-0 itsfPe sc-bdVaJa bxVNCd'):
    if 'm²' in str(item):
        if(item.text):
            metros.append(item)
        else:
            metros.append('')
```

```python
quartos = []


itens = site.find('ul', id='ad-list')

for item in itens.find_all('span', class_='sc-1ftm7qz-0 itsfPe sc-bdVaJa bxVNCd'):
  if 'm²' in str(item):
    if(item.text):
            metros.append(item)
    
```

Depois de extraídas todas as informações, utrilizando a biblioteca pandas criamos o DataFrame que seria uma tabela para melhor vizualização dos dados indexando  e ordenando em colunas.

```python
df = pd.DataFrame(list(zip(titulos, links, valores, localizacao, data, metros, quartos, cond)),
                  columns=['titulo', 'link', 'valores', 'localização', 'data de postagem', 'm²', 'qnt de quartos', 'valor do condominio' ])
```
